## Preprocesamiento de los datos para Machine Learning

Aplicar las técnicas principales de preprocesamiento de datos para preparar un 
dataset antes del entrenamiento.

## Parte 1. Compresión del problema

Trabajaremos con el Dataset de Titanic, para construir un modelo que sea capaz de 
predecir si una persona sobrevive o no.

```python
# importamos las librerias
import pandas as pd
import numpy as np

#Librerias de Machine Learning
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

#Librerias para graficar
import seaborn as sns
import matplotlib.pyplot as plt

#Libreria de Kaggle
import kagglehub
```

```python
#Descargamos el Dataset de la pagina Kaggle
path = kagglehub.dataset_download("yasserh/titanic-dataset")
path #Imprimimos la ruta
```

```python
dfTitanic = pd.read_csv(path + "/Titanic-Dataset.csv") #Creamos el Dataframe
```

```python
dfTitanic.head(5)
```

```python
dfTitanic.info()
```

## Parte 2. Selección de Variables Predictoras y Variable Objetivo

La variable objetivo sera : Survived
donde 0 = No sobrevivio 1 = Sobrevivio

```python
dfTitanic["Survived"].value_counts()
```

```python
dfTitanic.groupby(
    "Sex"
)["Survived"].mean()
```

```python
dfTitanic.groupby(
    "Pclass"
)["Survived"].mean()
```

```python
dfTitanic.groupby(
    "Fare"
)["Survived"].mean()
```

## Parte 3. Correlación e Hipótesis

Vamos a identificar la relación entre variables númericas.

nota: La correlación NO representa la importancia REAL de la variable en el modelo.

```python
variables_numericas = [
    "Survived",
    "Pclass",
    "SibSp",
    "Parch",
    "Fare",
]
```

```python
corr = dfTitanic[variables_numericas].corr() #Creamos la matriz de correlación
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)
plt.title("Matriz de correlación")
plt.show()
```

## Parte 4. Variables predictoras.

X --> Aquellas variables que nos ayudan a predecir

y --> Variable objetivo

```python
columnas= [
    "Pclass",
    "Sex",
    "Age",
    "Fare",
    "Embarked"
]
```

```python
X = dfTitanic[columnas]
y = dfTitanic["Survived"]
```

```python
X.columns
```

```python
y.name
```

## Parte 5. Limpieza de los datos

Realizaremos la imputación de los datos, donde existan valores nulos.

```python
# Revisamos si existen valores nulos
X.isnull().sum()
```

```python
#Vamos a imputar los datos para el campo Edad
X["Age"] = X["Age"].fillna(
    X["Age"].median()
)
```

```python
#Vamos imputar los datos para el campo Embarked, calculando la moda
X["Embarked"] = X["Embarked"].fillna(
    X["Embarked"].mode()[0]
)
```

```python
X.isnull().sum()
```

## Parte 6: Normalizamos los datos categoricos (texto)

Los algoritmos de ML no entienden texto.

```python
X.head(2)
```

```python
#One Hot Encoding
X= pd.get_dummies(
    X,
    columns=[
        "Sex",
        "Embarked"
    ],
    drop_first=True
)
```

```python
X.head(2)
```

## Parte 7. División de los datos

No debemos entrenar y evaluar utilizando los mismos datos.

80% para entrenar

20% para evaluar

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("Entrenamiento X: ", X_train.shape)
```

```python
print("Prueba X: ", X_test.shape)
```

```python
print("Entrenamiento Y: ", y_train.shape)
```

```python
print("Prueba Y: ", y_test.shape)
```

## Escalado de los datos
Las variables númericas tienen valores distintos. El escalado evita que las variables con mayores
valores dominen el aprendizaje del modelo.

```python
columnas_numericas = [
    "Age",
    "Fare"
]
```

```python
X_train[columnas_numericas].describe()
```

```python
# Escalamos los datos usando StandarScaler
scaler = StandardScaler()
```

```python
X_train[columnas_numericas] = scaler.fit_transform(X_train[columnas_numericas])
X_test[columnas_numericas] = scaler.transform(X_test[columnas_numericas])
```

```python
X_train[columnas_numericas].describe()
```

## Entrenamiento del Algoritmo:Arbol de decisión

Permite clasificar los registros con respecto a las variables

```python
modelo = DecisionTreeClassifier(
    max_depth=4,
    random_state=42
)
```

```python
modelo.fit(X_train, y_train)
```

## Evaluamos la eficacia del modelo

Realizamos predicciones sobre el conjunto de prueba

```python
# Invocamos el modelo el metodo predict con los datos de prueba no vistos.
prediccion = modelo.predict(
    X_test
)
```

```python
pd.DataFrame(
    {
        "Survived Real: " : y_test,
        "Survived Predicha: " : prediccion
    }
).head(20)
```

## Evaluación del modelo

Utilizaremos el accuracy para determinar el porcentaje de predicciones correctas realizadas por el modelo.

```python
accuracy = accuracy_score(y_test, prediccion)
```

```python
print(f"La exactitud del modelo es: {accuracy: .4f}")
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, prediccion))
```
