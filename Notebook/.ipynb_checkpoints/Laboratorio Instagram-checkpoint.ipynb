```python
#Importar las librerias necesarias
import pandas as pd
import random
```

```python
#Conjunto de datos base para la generación del Big Data
#Datos simulados de influencers de turismo, playas y lifestyle de Cancún
datos = {
    "influencer": ["cancun_vibes", "playa_lover", "turismo_mx", "lifestyle_cancun", "beach_queen", "sol_y_mar"],
    "likes": [1500, 3200, 870, 4100, 2750, 980],
    "comentarios": [120, 340, 55, 520, 210, 70],
    "fecha": ["2026-01-15", "2026-02-20", "2026-03-05", "2026-04-10", "2026-05-18", "2026-06-01"],
    "tipo_post": ["foto", "video", "reel", "foto", "reel", "video"]
}
```

```python
#Creamos el dataframe base
dfDatos = pd.DataFrame(datos)
```

```python
#Mostramos los datos base
dfDatos
```

```python
#Cantidad de registros a generar aleatoriamente
cantidad = 1000000
```

```python
#Crear el dataFrame de 1000000 de registros
dfMillon = pd.DataFrame({
    "influencer": [random.choice(dfDatos["influencer"]) for _ in range(cantidad)],
    "likes": [random.randint(100, 50000) for _ in range(cantidad)],
    "comentarios": [random.randint(0, 5000) for _ in range(cantidad)],
    "fecha": [random.choice(dfDatos["fecha"]) for _ in range(cantidad)],
    "tipo_post": [random.choice(dfDatos["tipo_post"]) for _ in range(cantidad)]
})
```

```python
#Verificamos la forma del dataframe (filas, columnas)
dfMillon.shape
```

```python
#Verificamos el uso de memoria del dataframe
dfMillon.memory_usage(deep=True)
```

```python
#Uso de memoria total en MB
dfMillon.memory_usage(deep=True).sum() / 1024**2
```

```python
#Verificamos los tipos de datos de cada columna
dfMillon.dtypes
```

```python
#Mostramos las primeras filas del dataset grande
dfMillon.head(10)
```

```python
#Insight 1: Promedio de likes y comentarios por influencer
dfMillon.groupby("influencer")[["likes", "comentarios"]].mean().round(2)
```

```python
#Insight 2: Total de publicaciones por tipo de post
dfMillon["tipo_post"].value_counts()
```

```python
#Insight 3: Influencer con mayor promedio de engagement (likes + comentarios)
dfMillon["engagement"] = dfMillon["likes"] + dfMillon["comentarios"]
dfMillon.groupby("influencer")["engagement"].mean().sort_values(ascending=False)
```
